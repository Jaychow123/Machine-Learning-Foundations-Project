```python
#Importing required libraries
import numpy as np
import pandas as pd
```

```python
#Loading data set
from sklearn import datasets
iris = datasets.load_iris()
```

```python
 #visualization of Iris dataset features using Matplotlib, with colors representing different classes and labeled axes
import matplotlib.pyplot as plt
_, ax = plt.subplots()
scatter = ax.scatter(iris.data[:, 0], iris.data[:, 1], c=iris.target)
ax.set(xlabel=iris.feature_names[0], ylabel=iris.feature_names[1])
_ = ax.legend(
    scatter.legend_elements()[0], iris.target_names, loc="lower right", title="Classes"
)
```

```python
# reducing its dimensionality to 3 dimensions, and then visualizes the data points in a 3D scatter plot
import mpl_toolkits.mplot3d
from sklearn.decomposition import PCA
fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d", elev=-150, azim=110)
X_reduced = PCA(n_components=3).fit_transform(iris.data)
ax.scatter(
    X_reduced[:, 0],
    X_reduced[:, 1],
    X_reduced[:, 2],
    c=iris.target,
    s=40,
)
ax.set_title("First three PCA dimensions")
ax.set_xlabel("1st Eigenvector")
ax.xaxis.set_ticklabels([])
ax.set_ylabel("2nd Eigenvector")
ax.yaxis.set_ticklabels([])
ax.set_zlabel("3rd Eigenvector")
ax.zaxis.set_ticklabels([])
plt.show()
```

# **2.SVM Implementation**

```python
#importing required libraries and evalution metrices
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
#Splitting the 80% data for training and 20% data for testing
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2, random_state=42)
```

```python
#Defining the model using Linear kernel
model1=SVC(kernel='linear')
model1.fit(x_train,y_train)
```

```python
#predicting the results
y_pred=model1.predict(x_test)
```

```python
#printing the accuracy score
print(accuracy_score(y_test,y_pred))
```

```python
#Defining the model using polynomial kernel
model2=SVC(kernel='poly')
model2.fit(x_train,y_train)
```

```python
#predicting the results
y_pred=model2.predict(x_test)
```

```python
#finding the accuracy
print(accuracy_score(y_test,y_pred))
```

```python
#Defining the model using rbf kernel
model3=SVC(kernel='rbf')
model3.fit(x_train,y_train)
```

```python
#predicting the result
y_pred=model3.predict(x_test)
```

```python
#finding the accuracy
print(accuracy_score(y_test,y_pred))
```

```python
#Defining the model using sigmoid kernel
model4=SVC(kernel='sigmoid')
model4.fit(x_train,y_train)
```

```python
#model prediction
y_pred=model4.predict(x_test)
```

```python
#printing accuracy
print(accuracy_score(y_test,y_pred))
```

# **3.K-fold Cross-Validation**

```python
from sklearn.model_selection import cross_val_score
```

Here we are considering rbf kernel over linear because it is impossible to get the accuracy 100% with the linear kernel for this dataset.Getting 100% accuracy for linear kernel indicates the model is overfitted and also it is impossible to seperate the dataset linearly.

```python
#Defining th
model5= SVC(kernel='rbf')
```

```python
#Defining the number of folds
k = 5
scores = cross_val_score(model5, X, Y, cv=k)
```
